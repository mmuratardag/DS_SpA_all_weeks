```python
import pandas as pd
gm_3v = pd.read_excel('gapminder_lf_1800_2015.xlsx', index_col=0).sort_values(by=['country', 'continent', 'year'])
gm_3v.head()
```

```python
gm_3v.tail()
```

```python
gm_1v = pd.read_excel('gapminder_plotly_lf_1957_2007.xlsx', index_col=0).sort_values(by=['country', 'continent', 'year'])
gm_1v.head()
```

```python
gm_1v.tail()
```

```python
df = pd.merge(left = gm_1v, right = gm_3v, on = ['country', 'continent', 'year'], how = 'inner')
df.head()
```

```python
df = df[['country','continent','year','gdpPercap','fertility_rate','total_population','life_expectancy']]
df.head()
```

```python
round(df.describe(), 3)
```

```python
import plotly.express as px
```

```python
fig_life = px.scatter(df, x="gdpPercap", y="life_expectancy", animation_frame="year", animation_group="country",
                      size="total_population", color="continent", hover_name="country", 
                      log_x = True, 
                      size_max=45, range_x=[100,100000], range_y=[10,90])
```

```python
fig_life.write_html('gm_px_life.html')
```

```python
fig_fert = px.scatter(df, x="gdpPercap", y="fertility_rate", animation_frame="year", animation_group="country",
                      size="total_population", color="continent", hover_name="country", 
                      log_x = True, 
                      size_max=45, range_x=[100,100000], range_y=[1,10])
```

```python
fig_fert.write_html('gm_px_fert.html')
```
