```python
import requests
import re

from bs4 import BeautifulSoup
import pandas as pd

import time
```

```python
sy = requests.get('https://www.lyrics.com/artist/Sonic-Youth/5474').text
pa = requests.get('https://www.lyrics.com/artist/Pavement/23382').text
dj = requests.get('https://www.lyrics.com/artist/Dinosaur-Jr./4096').text
se = requests.get('https://www.lyrics.com/artist/Sebadoh/14199').text
```

```python
sy_soup = BeautifulSoup(sy, 'html.parser')
# print(sy_soup.prettify())

"""

<a href="/lyric/9912643/Sonic+Youth/Razor+Blade">

<a href="/lyric/34538035/Sonic+Youth/Eric%27s+Trip">

<a href="/lyric/34538036/Sonic+Youth/Candle">

#####################################################################

<td class="tal qx">
             <strong>
              <a href="/lyric/27294305/Sonic+Youth/Screaming+Skull">
               Screaming Skull
              </a>
             </strong>
            </td>
            <td class="tar qx">
            
#####################################################################

""";
```

```python
# sonic youth

sy_lyrics = []
tables = sy_soup.find_all(class_='tdata')
for t in tables:
    cells = t.find_all('td', class_="tal qx")
    for c in cells:
        links = c.find_all('a')
        for l in links:
            sy_lyrics.append((l['href']))
```

```python
# sonic youth
sy_df_lyrics = pd.DataFrame(sy_lyrics)
sy_df_lyrics = sy_df_lyrics[sy_df_lyrics[0].str.contains('lyric')]
sy_df_lyrics['url'] ='https://www.lyrics.com'+sy_df_lyrics[0]
sy_df_lyrics.head()
```

```python
# pavement

pa_soup = BeautifulSoup(pa, 'html.parser')

pa_lyrics = []
tables = pa_soup.find_all(class_='tdata')
for t in tables:
    cells = t.find_all('td', class_="tal qx")
    for c in cells:
        links = c.find_all('a')
        for l in links:
            pa_lyrics.append((l['href']))
            
pa_df_lyrics = pd.DataFrame(pa_lyrics)
pa_df_lyrics = pa_df_lyrics[pa_df_lyrics[0].str.contains('lyric')]
pa_df_lyrics['url'] ='https://www.lyrics.com'+pa_df_lyrics[0]
pa_df_lyrics.head()
```
