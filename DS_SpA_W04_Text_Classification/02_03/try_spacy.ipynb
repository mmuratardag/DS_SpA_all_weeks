```python
import pandas as pd

import spacy
nlp = spacy.load('en_core_web_md')

from dframcy import DframCy
dframcy = DframCy(nlp)
```

```python
corpus_df = pd.read_csv('corpus_df.csv').drop(columns=['Unnamed: 0'])
corpus_df.head()
```

```python
sy_rnd_son_lyr = str(corpus_df[corpus_df['artist'] == 'sonic youth']['lyrics'].sample(n=1, random_state=666).values[0])
st_rnd_son_lyr = str(corpus_df[corpus_df['artist'] == 'stooges']['lyrics'].sample(n=1, random_state=666).values[0])
```

```python
d_sy_rnd_son_lyr = dframcy.nlp(sy_rnd_son_lyr)
sy_s_df = dframcy.to_dataframe(d_sy_rnd_son_lyr)
sy_s_df
```

```python
spacy.displacy.render(nlp(sy_rnd_son_lyr), style="ent")
```

```python
d_st_rnd_son_lyr = dframcy.nlp(st_rnd_son_lyr)
st_s_df = dframcy.to_dataframe(d_st_rnd_son_lyr)
st_s_df
```

```python
spacy.displacy.render(nlp(st_rnd_son_lyr), style="ent")
```
