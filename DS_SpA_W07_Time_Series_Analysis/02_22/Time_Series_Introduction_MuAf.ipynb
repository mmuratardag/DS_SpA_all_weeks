How is a Time Series problem different from a "classic" regression problem?
- there will be something called auto-correlation. Some variable will correlate with itself.
- In Time Series, we are actually using as inputs (X) the same variable that we are predicting (y), where the variables are just shifted back in time.
- X:
    - temperature from 1 day ago,
    - temperature from 2 days ago,
    - temperature from N days ago.
- y = temperature today.

Time Series anlaysis is a field of "machine learning" / statistical analysis which models something called "time dependence".

General Tip / Best Practice:
- When dealing with Time Series data in Pandas, convert the time / data to Timestamps!!!
    - Will make your life MUCH easier.

```python
pd.to_datetime('2 Feb 2021')
```

```python
df['date'] = df['year'].astype(str) + '-' + df['month'].astype(str)
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.set_index('date', inplace=True)
```

```python
df.drop(['year', 'month'], axis=1, inplace=True)
```

```python
# df['Jan 1950': 'February 1954'].plot()
sns.set()
df.plot(figsize=(12, 6))
```

### 3 Components of Time Series Data

1. Trend
    - Linear
    - Polynomial
2. Seasonality
    - Daily
    - Monthly
    - Quarterly
    - Yearly
    - Any fixed interval
3. Remainder (Once Trend and Seasonality are Removed, We are left with a so-called "stationary" time series)
    - Time Dependence <---- **THIS IS WHAT TIME SERIES ANALYSIS IS ALL ABOUT!!!**
    - Random Noise
