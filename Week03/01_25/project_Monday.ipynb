```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('train.csv', parse_dates=True)
train.head()
```

```python
train.dtypes
```

```python
train['datetime'] = pd.to_datetime(train['datetime'])
train.head()
```

```python
train.dtypes
```

```python
train['year'] = train['datetime'].dt.year
train['month'] = train['datetime'].dt.month
train['day']= train['datetime'].dt.day
train.head()
```

```python
round(train.describe().T, 2)
```

```python
train_s1 = train[train['season']==1]
train_s1[['casual', 'registered']].plot(figsize=(15, 10), alpha=.4)
```

```python
train_s2 = train[train['season']==2]
train_s2[['casual', 'registered']].plot(figsize=(15, 10), alpha=.4)
```

```python
train_s3 = train[train['season']==3]
train_s3[['casual', 'registered']].plot(figsize=(15, 10), alpha=.4)
```

```python
train_s4 = train[train['season']==4]
train_s4[['casual', 'registered']].plot(figsize=(15, 10), alpha=.4)
```

```python
train[['casual', 'registered', 'season']].set_index('season').plot(figsize=(15, 10), alpha=.4)
```
